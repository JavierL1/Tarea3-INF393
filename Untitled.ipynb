```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

print("UWU")
```

```python
# Read data
dftrain = pd.read_csv('bike_sharing_train.csv')
dfval = pd.read_csv('bike_sharing_val.csv')
dftest = pd.read_csv('bike_sharing_test.csv')

ntrain = len(dftrain)
nval = len(dftrain) + len(dfval)

df = pd.concat([dftrain,dfval,dftest])
print ('\nSummary - dataframe completo:\n')
print (df.describe())
df['hour'] = pd.to_datetime(df['datetime']).apply(lambda x: x.strftime('%H'))
df['hour'] = pd.to_numeric(df['hour'])
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel("Hora del da", fontsize=16)
plt.ylabel("Arriendos Totales", fontsize=16)

# Bicicletas en el verano
dfsum = df.loc[df['season'] == 2]
plt.plot(dfsum['hour']-0.3, dfsum['count'], 'o', color=(1,0,0), alpha=1, label="Verano")

# Bicicletas en el otoo
dfaut = df.loc[df['season'] == 3]
plt.plot(dfaut['hour']-0.1, dfaut['count'], 'o', color=(0.2,1,1), alpha=1, label="Otoo")

# Bicicletas en el invierno
dfwin = df.loc[df['season'] == 4]
plt.plot(dfwin['hour']+0.1, dfwin['count'], 'o', color=(0,0,1), alpha=1, label="Invierno")

# Bicicletas en el primavera
dfspr = df.loc[df['season'] == 1]
plt.plot(dfspr['hour']+0.3, dfspr['count'], 'o', color=(0,1,0), alpha=1, label="Primavera")

plt.title('Cantidad de arriendos de bicicletas por hora del da', fontsize=18)
plt.legend(fontsize=16, title='Estaciones del ao')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel("Hora del da", fontsize=16)
plt.ylabel("Arriendos Totales", fontsize=16)

# Bicicletas en clima despejado
dfdes = df.loc[df['weather'] == 1]
plt.plot(dfdes['hour']-0.3, dfdes['count'], 'o', color=(0,1,0), alpha=1, label="Despejado")

# Bicicletas en clima con neblina
dfneb = df.loc[df['weather'] == 2]
plt.plot(dfneb['hour']-0.1, dfneb['count'], 'o', color=(0.2,1,1), alpha=1, label="Neblina")

# Bicicletas en clima de lluvia ligera
dflll = df.loc[df['weather'] == 3]
plt.plot(dflll['hour']+0.1, dflll['count'], 'o', color=(1,0.4,0), alpha=1, label="Lluvia Ligera")

# Bicicletas en clima de lluvia intensa y nieve
dfllf = df.loc[df['weather'] == 4]
plt.plot(dfllf['hour']+0.3, dfllf['count'], 'o', color=(1,0,0), alpha=1, label="Lluvia Pesada/Nieve")

plt.title('Cantidad de arriendos de bicicletas por hora del da', fontsize=18)
plt.legend(fontsize=16, title='Estaciones del ao')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel("Hora del da", fontsize=16)
plt.ylabel("Arriendos Totales", fontsize=16)

# Bicicletas en da de trabajo
dftra = df.loc[df['workingday'] == 1]
plt.plot(dftra['hour']-0.3, dftra['count'], 'o', color=(0,0,1), alpha=1, label="Trabajo")

# Bicicletas en fines de semana y vacaciones
dfvac = df.loc[df['workingday'] == 0]
plt.plot(dfvac['hour']-0.1, dfvac['count'], 'o', color=(0.2,1,1), alpha=1, label="Fin de Sem. o Vacaciones")

plt.title('Cantidad de arriendos de bicicletas por hora del da', fontsize=18)
plt.legend(fontsize=16, title='Estaciones del ao')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
# plt.xlabel("Grados Celsius", fontsize=16)
plt.ylabel("Grados Celsius", fontsize=16)

# Bicicletas en da de trabajo
plt.plot(range(len(df['temp'])), df['temp']-df['atemp'], '-', color=(0,0,1), alpha=1, label="Trabajo")

plt.title('Diferencia entre temperatura real y sensitiva', fontsize=18)
plt.legend(fontsize=16, title='Diferencia')
plt.show()
```
